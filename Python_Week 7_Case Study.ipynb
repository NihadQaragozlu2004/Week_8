```python
columns =data.columns[1:]
fig,axes = plt.subplots(nrows=len(columns), ncols=1, figsize=(9,25))
for i in range(len(columns)) :
    sns.boxplot(ax=axes[i], y=columns[i], data=data)
```

```python
def outlier_treatment(data):
    q1,q3 = data.quantile([.25, .75])
    iqr = q3 - q1
    lower_range = q1 - (1.5 * iqr)
    upper_range = q3 + (1.5 * iqr)   
    return lower_range , upper_range
```

```python
for col in data.columns[1:]:
    lowerbound, upperbound = outlier_treatment(data[col])
    data[col] = np.clip(data[col], a_min=lowerbound, a_max=upperbound)
```

```python
columns =data.columns[1:]
fig,axes = plt.subplots(nrows=len(columns), figsize=(9,25))
for i in range(len(columns)) :
    sns.boxplot(ax=axes[i], y=columns[i], data=data)
```

```python
from sklearn.preprocessing import StandardScaler

X = data.iloc[:, 1:]
cols = data.columns[1:]
index = data['country']
scaler = StandardScaler()
X = scaler.fit_transform(X)
scale_data = pd.DataFrame(X, columns=cols, index=index)
scale_data.head()
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(scale_data)
```

```python
plt.bar(height=pca.explained_variance_, x=range(1, len(pca.explained_variance_)+1))
```

```python
pca = PCA(n_components=5)
scale_data_pca = pd.DataFrame(pca.fit_transform(scale_data), index=index)
scale_data_pca.head()
```

```python
scale_data_pca = scale_data
```

# With K-Means

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters=3, random_state=123)
k_means_pred = k_means.fit_predict(scale_data_pca)
k_means_pred
```

```python
data['K_Means_Clusters'] = k_means_pred
```

```python
data['K_Means_Clusters'].value_counts()
```

```python
plt.figure(figsize=(9,6), dpi=100)
sns.scatterplot(data=data, x='income', y='gdpp', hue='K_Means_Clusters', palette="coolwarm")
plt.grid()
plt.show()
```

```python
data['K_Means_Clusters'].replace({0:'Developed', 1:'Developing', 2:'Underdeveloped'}, inplace=True)
```

```python
plt.figure(figsize=(9,6), dpi=100)
sns.scatterplot(data=data, x='income', y='gdpp', hue='K_Means_Clusters', palette="husl")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9,6), dpi=100)
sns.scatterplot(data=data, x='health', y='life_expec', hue='K_Means_Clusters', palette="husl")
plt.grid()
plt.show()
```

# Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage,dendrogram
```

```python
plt.figure(figsize=(9,6), dpi=100)
dendrogram(linkage(scale_data_pca, method='ward', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
hc_pred = hc.fit_predict(scale_data_pca)
hc_pred
```

```python
data['HC_Clusters'] = hc_pred
```

```python
data.head()
```

```python
plt.figure(figsize=(9,6), dpi=100)
sns.scatterplot(data=data, x='income', y='gdpp', hue='HC_Clusters', palette="coolwarm")
plt.show()
```

```python
data['HC_Clusters'].replace({0:'Developed', 1:'Developing', 2:'Underdeveloped'}, inplace=True)
```

```python
plt.figure(figsize=(9,6), dpi=100)
sns.scatterplot(data=data, x='income', y='gdpp', hue='HC_Clusters', palette="coolwarm")
plt.show()
```

```python
plt.figure(figsize=(9,6), dpi=100)
sns.scatterplot(data=data, x='health', y='life_expec', hue='HC_Clusters', palette="husl")
plt.show()
```
